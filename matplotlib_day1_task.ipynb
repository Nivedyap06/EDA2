```python
#	Task 1: Basic Line Plot
#	Create a line graph that shows the relationship between x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10].
# Label the axes as "X-axis" and "Y-axis".
#	Set the title to "Basic Line Plot".
# Hint: Use plt.plot() and plt.show().


import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5]) #x points
y=np.array([2,4,6,8,10]) #y points

plt.xlabel("X axis") #x label
plt.ylabel("y axis")# y label
plt.title("Basic Line Plot") # set title

plt.plot(x,y,color='black')
plt.show() # display
```

```python
# Task 2: Line Graph with Markers
#	Create a line graph for x = [0, 1, 2, 3, 4] and y = [0, 1, 4, 9, 16].
#	Add circular markers to the points and connect them with a line.
#	Set the title as "Line Plot with Markers".
#Hint: Use plt.plot(x, y, marker='o').

import matplotlib.pyplot as plt
import numpy as np
x= np.array([0,1,2,3,4])
y=np.array([0,1,4,9,16])
plt.title("Line Plot with Markers")
plt.plot(x, y,'o',color='maroon')
plt.show()
```

```python
# Task 3: Multiple Line Plots
#	Plot two lines on the same graph:
# Line 1: x = [1, 2, 3, 4] and y1 = [1, 4, 9, 16]
# Line 2: x = [1, 2, 3, 4] and y2 = [2, 3, 6, 8]
# Use different colors and line styles for each line (e.g., one dashed line and one solid line).
# Add a legend to differentiate the lines.
# Hint: Use plt.plot() twice and add plt.legend().


import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4])
y=x**2
plt.legend(['single element'])
plt.plot(x,y,linestyle='dashed',color='r')
plt.show()


x=np.array([1,2,3,4])
y=np.array([2,3,6,8])
plt.legend(['solid element'])
plt.plot(x,y,linestyle='solid',color='g')
plt.show()

```

```python
#Task 4: Subplots with Line Graphs
#Create two separate line graphs in one figure:
#first graph: x = [1, 2, 3, 4, 5] and y = [1, 4, 9, 16, 25]
#Second graph: x = [1, 2, 3, 4, 5] and y = [2, 4, 6, 8, 10]
#Use plt.figure() and plt.subplot() to place the graphs side by side.
#Use plt.figure() to create a new figure, and plt.subplot(1, 2, 1) for the first plot and plt.subplot(1, 2, 2) for the second.

import matplotlib.pyplot as plt
import numpy as np
plt.figure (figsize=(10,5))

# first subplot
plt.subplot(1,2,1)
x1=np.array([1, 2, 3, 4, 5])
y1=np.array([1, 4, 9, 16, 25])
plt.plot(x1,y1,color='pink')
plt.title("first graph")

# second subplot
plt.subplot(1,2,2)
x2=np.array([1, 2, 3, 4, 5])
y2=np.array([2,4,6,8,10])
plt.plot(x2,y2,color='orange')
plt.title("second graph")

plt.figure()



```

```python
# Task 5 Basic Bar Plot
# 	Create a bar plot with categories ['A', 'B', 'C', 'D'] and values [5, 7, 3, 8].
# 	Label the x-axis as "Categories" and the y-axis as "Values".
# oAdd a title: "Basic Bar Plot".
# Hint: Use plt.bar() for bar plots.


import matplotlib.pyplot as plt
import numpy as np
x=np.array(["A","B","C","D"])
y=np.array([5,7,3,8])
plt.title(" Basic Bar Plot")
plt.bar(x,y, color='indigo')
plt.show()
```

```python
	#Task 6: Horizontal Bar Plot
#Create a horizontal bar plot for categories ['E', 'F', 'G', 'H'] and values [8, 4, 6, 3].
#Set the title to "Horizontal Bar Plot".
# Use plt.barh() for horizontal bar plots.


import matplotlib.pyplot as plt
import numpy as np

x=np.array(['E', 'F', 'G', 'H'])
y=np.array([8, 4, 6, 3])
plt.title("Horizontal Bar Plot")
plt.barh(x,y,color='red')
plt.show()
```

```python

#	Task 8: Create Two Figures with Different Plots
# 	In one figure, create a line plot for x = [1, 2, 3, 4] and y = [1, 4, 9, 16].
# In a second figure, create a bar plot for labels = ['A', 'B', 'C'] and values = [5, 7, 9].
# Hint: Use plt.figure() twice to create separate figures.


import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=x**2
plt.plot(x,y,color='grey')
plt.figure()

x1=np.array(['A','B','C'])
y1=np.array([5,7,9])
plt.bar(x1,y1,color='brown')
plt.figure()
```

```python
#	Task 9: Adjust Figure Size
#	Create a line plot for x = [1, 2, 3, 4, 5] and y = [5, 3, 8, 7, 4].
#	Change the size of the figure to 10x6 inches using plt.figure(figsize=(10, 6)).
# Add appropriate labels and a title.
# Pass the figsize parameter inside plt.figure().
import matplotlib.pyplot as plt
import numpy as np

x=np.array([1,2,3,4,5])
y=np.array([5,3,8,7,4])
plt.figure(figsize=(10,6))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title(" figure size")
plt.plot(x,y,color='yellow')
plt.show()
```
